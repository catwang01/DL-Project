# Tensorflow 获取所有 tensor

## 从 graph_def 中获取

```python
import tensorflow as tf

tf.reset_default_graph()

# 从 meta 中加载 graph
saver = tf.train.import_meta_graph('cased_L-12_H-768_A-12/bert_model.ckpt.meta')
graph = tf.get_default_graph()

# 从 graph 中获取 graph_def
for node in graph.as_graph_def().node:
    print(node.name)
```
