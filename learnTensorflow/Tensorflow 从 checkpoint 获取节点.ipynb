# Tensorflow 从 checkpoint 获取节点

我们使用 bert 的模型作为示例，下载地址 [https://storage.googleapis.com/bert_models/2018_10_18/cased_L-12_H-768_A-12.zip](https://storage.googleapis.com/bert_models/2018_10_18/cased_L-12_H-768_A-12.zip)

```python
!ls cased_L-12_H-768_A-12
```

```python
import tensorflow as tf

saver = tf.train.import_meta_graph('./cased_L-12_H-768_A-12/bert_model.ckpt.meta')j
```

此时，我们已经将图结构导入到默认的 graph 中了，我们获取默认 graph

```python
graph = tf.get_default_graph()
```

我们向看看 graph 中有多少节点，可以用下面的代码

```python
for node in graph.as_graph_def().node:
    print(node.name)
```

如果我们想获取 `bert/embeddings/word_embeddings:0` 这个节点，可以使用下面的代码

```python
embedding_table = graph.get_tensor_by_name("bert/embeddings/word_embeddings:0")
```

获取了这个节点后，就可以查看节点的值了，使用下面的代码

```python
with tf.Session() as sess:
    saver.restore(sess, './cased_L-12_H-768_A-12/bert_model.ckpt')
    print(sess.run(embedding_table))
```
